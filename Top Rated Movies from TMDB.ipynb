```python
from wordcloud import WordCloud
```

```python
# Converting feature to string
text = ' '.join(map(str, data.original_title.values))
```

```python
wordcloud = WordCloud(width=1100, height=1000, min_font_size=10, random_state=11).generate(text)

# Plotting WordCloud plot
plt.figure(figsize=(10,9))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
